```python
import pandas as pd
df=pd.read_csv("C:\\Users\\DELL\\Desktop\\odi.csv")
```

```python
df
```

```python
#Feature and Target
X = df.iloc[:,[7,8,9,10,11,12,13]].values
Y = df.iloc[:, 14].values
```

```python
#train-test splitting
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state =123)
```

```python
#Feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#Linear Regression
#statsmodel for constant and model OLS
import statsmodels.api as sm
#add constant
X_train_const=sm.add_constant(X_train)
model1=sm.OLS(Y_train, X_train_const).fit()
model1.summary()
```

```python
model1.params
```

```python
#using scikit learn- linear Regression
from sklearn.linear_model import LinearRegression
from sklearn import metrics
model2=LinearRegression()
model2.fit(X_train, Y_train)
Y_pred=model2.predict(X_test)
```

```python
import numpy as np
#means square error
rm=(np.sqrt(metrics.mean_squared_error(Y_test, Y_pred)))/100
accuracy=1-rm
print(accuracy)
#accuracy is around 57%
```

```python
#decision tree regression
from sklearn.tree import DecisionTreeRegressor  
regressor = DecisionTreeRegressor(random_state = 123)  
model3=regressor.fit(X_train, Y_train) 
Y_pred2=model3.predict(X_test)
rm2=(np.sqrt(metrics.mean_squared_error(Y_test, Y_pred2)))/100
accuracy2=1-rm2
print(accuracy2)
#accuracy is around 64%
```

```python
#random forest regression
from sklearn.ensemble import RandomForestRegressor 
regressor = RandomForestRegressor(n_estimators = 100, random_state = 0) 
model4=regressor.fit(X_train, Y_train) 
Y_pred3=model4.predict(X_test)
rm3=(np.sqrt(metrics.mean_squared_error(Y_test, Y_pred3)))/100
accuracy3=1-rm3
print(accuracy3)
#accuracy is around 72%
#best performance by random forest
```
